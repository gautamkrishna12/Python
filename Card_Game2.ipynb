```python
suits=['Hearts','Diamonds','Spades','Clubs']
ranks=['Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace']
values={'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':11,
         'Queen':12,'King':13,'Ace':14}
```

```python
class Cards():
    
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
        
    def __str__(self):
        return f"{self.suit} : {self.value}"
```

```python
class Deck():
    
    def __init__(self):
        self.Total_cards=[]
        for suit in suits:
            for rank in ranks:
                card=Cards(suit,rank)
                self.Total_cards.append(card)
    
    def Take_one_card(self):
        return self.Total_cards.pop(0)
    
    def Shuffle(self):
        from random import shuffle
        shuffle(self.Total_cards)
        
```

```python
class Player():
    def __init__(self,name):
        self.p_name=name
        self.p_cards=[]
    
    def Take_a_card(self):
        return self.p_cards.pop(0)
    
    def Add_a_Card(self,new_cards):
        if(type(new_cards)==type([])):
            self.p_cards.extend(new_cards)
        else:
            self.p_cards.append(new_cards)
            
    def __str__():
        return f"{self.p_name} has {len(p_cards)} Cards"
        
```
