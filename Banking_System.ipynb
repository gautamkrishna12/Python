```python
#Stimulating a user-friendly banking system that enables individuals to effortlessly open accounts 
#and perform hassle-free transactions, including withdrawals and deposits.
myaccounts=[]
accounts=[]
```

```python
class BankAccount():
    bank_name='Python_Bank'
    balance=0
    def __init__(self,name,acc,age,dob,ph,pas):
        self.name=name
        self.account_no=acc
        self.age=age
        self.dob=dob
        self.phone_no=ph
        self.password=pas
        
    def DisplayAccount(self):
        print('*'*26)
        print(f' Name       : {self.name}')
        print(f' Account_no : {self.account_no}')
        print(f' Phone_no   : xxxxxx{str(self.phone_no)[6::]}')
        print(f' Balance    : {self.balance}')
        print('*'*26)
        
    def Deposite(self):
        amount=float(input('Enter the amount : '))
        self.balance+=amount
        print('******Amount Successfully added to your account.******\n')
        
    def Withdraw(self):
        amount=float(input('Enter the amount : '))
        if(self.balance>amount):
            self.balance-=amount
            print('*******Collect your cash.********\n')
        else:
            print('Unsufficient Balance!')
        
    def __str__(self):
        return f'Name : {self.name}\nAccount_no : {self.account_no}'
    
    def __len__(self):
        return self.balance
```

```python
def Exit():
    print("Enter 'y' to continue ,'n' to exit")
    ch='p'
    while ch not in ['y','n']:
        ch=input('Enter your choice :')
        if(ch not in ['y','n']):
            print('Wrong Choice')
    if ch=='y':
        return True
    else:
        return False
```

```python
def getAccount_no():
    from random import randint
    while True:
        acc_no=randint(1000000000,9999999999)
        if acc_no not in accounts:
            accounts.append(acc_no)
            return acc_no
```

```python
def check_account(acc_no,pas):
    for i in myaccounts:
        if(i.account_no==acc_no and i.password==pas):
            return i
    else:
        return None
```

```python
def Choice():
    ch='p'
    while ch not in [1,2,3,4,5]:
        try:
            ch=int(input('Enter your choice :'))
            if(ch not in [1,2,3,4,5]):
                print('Wrong Choice')
        except:
            print('Enter only Digits')
    return ch
```

```python
def GetPass():
    import getpass
    Pass=getpass.getpass('Enter your password : ')
    return Pass
```
