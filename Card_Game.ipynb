```python
#Card Game
import random
suits=['Spades','Hearts','Diamonds','clubs']
ranks=['Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace']
values={'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,
        'Ten':10,'Jack':11,'Queen':12,'King':13,'Ace':14}
```

```python
class Card():
    #Creating cards class
    def __init__(self,c_name,rank):
        self.c_name=c_name
        self.c_rank=rank
        self.c_value=values[rank]
        
    def __str__(self):
        return f"{self.c_name} : {self.c_rank}"
```

```python
class Deck():
    
    def __init__(self):
        self.all_cards=[]
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
    
    def Shuffle(self):
        random.shuffle(self.all_cards)
    
    def Take_one_Card(self):
        return self.all_cards.pop()
```

```python
class Player():
    #Creating a class for players
    def __init__(self,name):
        self.p_name=name
        self.p_cards=[]
    
    def Take_a_Card(self):
        return self.p_cards.pop(0)
    
    def add_a_Card(self,new_card):
        if(type(new_card)==type([])):
            #if th player get multiple cards
            self.p_cards.extend(new_card)
        else:
            #if the player get only a card
            self.p_cards.append(new_card)
    
    def __str__(self):
        
        return f"{self.p_name} has {len(self.p_cards)}."
        
```

```python
rounds=1
game_ON=True
while game_ON:
    print(f"Round : {rounds}")
    rounds+=1
    
    if(len(player_one.p_cards)==0):
        print("Game over!!!")
        print("Player One Lost.\nPlayer Two Wins!")
        game_ON=False
        break
    
    if(len(player_two.p_cards)==0):
        print("Game over!!!")
        print("Player Two Lost.\nPlayer One Wins")
        game_On=False
        break
        
    #If the game is still going
    #Player One selets a card
    player1_cards=[]
    player1_cards.append(player_one.Take_a_Card())
    #PLayer two selects a card
    player2_cards=[]
    player2_cards.append(player_two.Take_a_Card())
    
    war_is_on=True
    while war_is_on:
        
        if( player1_cards[-1].c_value> player2_cards[-1].c_value):
            player_one.add_a_Card(player1_cards)
            player_one.add_a_Card(player2_cards)
            war_is_on=False
        
        elif(player1_cards[-1].c_value<player2_cards[-1].c_value):
            player_two.add_a_Card(player1_cards)
            player_two.add_a_Card(player2_cards)
            war_is_on=False
            
        else:
            print("***At WAR***")
            if(len(player_one.p_cards)<5):
                print("player1 out of cards!GAME OVER")
                print("Player1 Lost .PLayer 2 WINS.")
                war_is_on=False
                break
                
            elif (len(player_two.p_cards)<5):
                print("Player 2 out of cards.GAME OVER")
                print("Player2 loses .Player 1 wins.!")
                war_is_on=False
                break
            else:
                for i in range(5):
                    player1_cards.append(player_one.Take_a_Card())
                    player2_cards.append(player_two.Take_a_Card())
    
        
```
